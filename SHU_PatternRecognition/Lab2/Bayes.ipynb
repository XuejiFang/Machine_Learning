## 朴素贝叶斯
**优点**
小样本，多分类

**缺点**
输入数据敏感

**适用数据类型**
标称型数据

**公式**

$$
P(\omega_i|x)=\frac{P(\omega_i,x)}{p(x)}=\frac{P(\omega_i)p(x|\omega_i)}{\sum_{j}{P(\omega_j)p(x|\omega_j)}}
$$

对于同一个$x$，$p(x)$一样，因此只考虑分子部分

$$
h_{nb}=arg\ max_{c\in \gamma}P(c)\prod_{i=1}^d{P(x_i|c)}
$$

1. 离散
$$
p(c)=\frac{|D_c|}{|D|}
$$

> 拉普拉斯修正 $p(c)=\frac{|D_c|+1}{|D|+N}$


2. 连续

> 假设$p(x_i|c)\sim N(\mu_{c,i},\sigma_{c,i}^2)$

$$
p(x_i|c)=\frac{1}{\sqrt{2\pi}\sigma_{c,i}} exp(-\frac{(x_i-\mu_{c,i})^2}{2\sigma_{c,i}^2})
$$

```python
def load_data(fileName):
    data = pd.read_excel(fileName,index_col=0)
    data = data.dropna(axis=1)
    data['性别'] = data['性别'].map({'男':0,'女':1})
    sns.pairplot(data)
    return data
```

```python
def cal_prio(Label):
    """
    函数说明：
        计算先验概率
    parameters:
        - Label 标签列向量
    return:
        - prio 先验概率（字典）
    """
    from collections import Counter
    counts = Counter(Label)
    acc = 0
    for x in counts:
        acc += counts[x]
    prio = counts
    for x in prio:
        prio[x] /= acc
    return prio
```

```python
def cal_cond_discrete(Y, X):
    """
    函数说明：
        离散型数值，已知X下Y的概率
    parameters:
        - Y 目标概率
        - X 条件
    return:
        - Mat 概率矩阵，m*n
        - xLabel X的标签向量，1*m
        - yLabel Y的标签向量，1*n
    """
    from collections import Counter
    countX = Counter(X); countY = Counter(Y)
    
    xLabel = list(countX.keys()); yLabel = list(countY.keys())
    m = np.size(xLabel); n = np.size(yLabel); tol = np.size(X)
    countMat = np.zeros([m + 1, n])
    
    for index in range(tol):
        countMat[xLabel.index(X[index]), yLabel.index(Y[index])] += 1

    for j in range(n):
        countMat[-1:, j] = sum(countMat[:, j])
    for i in range(m):
        for j in range(n):
            # countMat[i, j] /= countMat[-1, j] 
            # 修正拉普拉斯
            countMat[i, j] = (countMat[i, j] + 1) / (countMat[-1, j] + m)
    return countMat[:-1, :], xLabel, yLabel
```

```python
def trainNB0(c, X, trainX):
    """
    函数说明：
        给定训练特征矩阵trainX，求X下，c的概率
    parameters:
        - c 训练标签
        - X 特征向量
        - trainX 训练特征矩阵
    return:
        - proDict 各标签的概率 
    """
    prio = cal_prio(c)
    m, n = np.shape(trainX)
    proMat = []
    yLabels = []
    for i in range(n):
        countMat, xLabel, yLabel = cal_cond_discrete(trainX[:, i], c)
        proMat.append(countMat)
        yLabels.append(yLabel)
    
    N = len(set(c)) # 训练集中的可能类别数
    
    proVec = []
    for cc in xLabel:
        p = prio[cc]
        index_c = xLabel.index(cc)
        for x in X:
            # 在cc下x的概率
            index_Class = list(X).index(x)
            try:
                index_x = yLabels[index_Class].index(x)
                p = p*proMat[index_Class][index_c, index_x]
            except ValueError:
                p = p*(1/(2*N))
        proVec.append(p)
    proDict = dict(zip(xLabel, proVec))
    return proDict
```
